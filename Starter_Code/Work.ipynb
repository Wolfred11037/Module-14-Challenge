```python
import requests
import json
```

```python

url = "https://2u-data-curriculum-team.s3.amazonaws.com/dataviz-classroom/v1.1/14-Interactive-Web-Visualizations/02-Homework/samples.json"

response = requests.get(url)

if response.status_code == 200:
    data = json.loads(response.text)
else:
    print("Failed to fetch data. Status code:", response.status_code)
```

```python
%%html
<div>
  <label for="selDataset">Select an individual:</label>
  <select id="selDataset"></select>
</div>

<div id="bar-chart"></div>

<script src="https://d3js.org/d3.v7.min.js"></script>
<script src="https://cdn.plot.ly/plotly-latest.min.js"></script>
<script>
  // JavaScript code for creating the dropdown menu and chart

  // Fetch the data from the JSON file
  d3.json("https://2u-data-curriculum-team.s3.amazonaws.com/dataviz-classroom/v1.1/14-Interactive-Web-Visualizations/02-Homework/samples.json").then(function(data) {

    // Extract necessary data
    var samples = data.samples;
    var sample_names = data.names;

    // Populate the dropdown with individual names
    var dropdown = d3.select('#selDataset');
    sample_names.forEach(name => {
      dropdown.append('option').text(name).attr('value', name);
    });

    // Function to update the chart based on selected individual
    function updateBarChart(selectedIndividual) {
      var selectedData = samples.find(sample => sample.id == selectedIndividual);
      var top10SampleValues = selectedData.sample_values.slice(0, 10).reverse();
      var top10OTUIds = selectedData.otu_ids.slice(0, 10).map(id => `OTU ${id}`);
      var top10OTULabels = selectedData.otu_labels.slice(0, 10);

      var trace = {
        x: top10SampleValues,
        y: top10OTUIds,
        text: top10OTULabels,
        type: "bar",
        orientation: "h"
      };

      var data = [trace];

      var layout = {
        title: `Top 10 OTUs for ${selectedIndividual}`,
        xaxis: { title: "Sample Values" },
        yaxis: { title: "OTU IDs" }
      };

      Plotly.newPlot("bar-chart", data, layout);
    }

    // Initial chart with default individual
    updateBarChart(sample_names[0]);

    // Event handler for dropdown change
    dropdown.on("change", function() {
      var selectedIndividual = dropdown.property("value");
      updateBarChart(selectedIndividual);
    });

  }).catch(function(error) {
    console.log("Error loading data:", error);
  });
</script>
```

```python
%%html
<div id="bubble-chart"></div>

<script>
  // JavaScript code for creating the bubble chart

  // Function to create/update the bubble chart
  function updateBubbleChart(selectedIndividual) {
    // Find the selected data for the individual
    var selectedData = samples.find(sample => sample.id == selectedIndividual);

    // Extract data for the bubble chart
    var otuIds = selectedData.otu_ids;
    var sampleValues = selectedData.sample_values;
    var otuLabels = selectedData.otu_labels;

    var trace = {
      x: otuIds,
      y: sampleValues,
      text: otuLabels,
      mode: 'markers',
      marker: {
        size: sampleValues,  // Use sample_values for marker size
        color: otuIds,       // Use otu_ids for marker colors
        colorscale: 'Viridis' // Choose a colorscale
      }
    };

    var data = [trace];

    var layout = {
      title: `Bubble Chart for ${selectedIndividual}`,
      xaxis: { title: 'OTU IDs' },
      yaxis: { title: 'Sample Values' },
      showlegend: false
    };

    Plotly.newPlot("bubble-chart", data, layout);
  }

  // Initial chart with default individual
  updateBubbleChart(sample_names[0]);

  // Event handler for dropdown change
  dropdown.on("change", function() {
    var selectedIndividual = dropdown.property("value");
    updateBubbleChart(selectedIndividual);
  });
</script>
```
